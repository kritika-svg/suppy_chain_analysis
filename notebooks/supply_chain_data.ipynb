# ***Data Loading & Cleaning***
**Import data, explore structure, and cleaning it**


```python
import warnings
warnings.filterwarnings('ignore')
# required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# for graph
plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")
```

```python
# dataset
df = pd.read_csv("supply_chain_data.csv")
print("✅ Data Loaded Successfully!")
df.head()
```

```python
print("----- Dataset Info -----")
df.info()

print("\n----- Missing Values -----")
print(df.isnull().sum())

print("\n----- Statistical Summary -----")
df.describe(include='all').T
```

`df.info() `→ Shows columns, data types, and missing values.

`df.describe()` → Gives numeric summary like mean, median, min, max.

```python
print("\n----- Product Type Distribution -----")
display(df['Product type'].value_counts())

print("\n----- Customer Demographics Distribution -----")
display(df['Customer demographics'].value_counts())
```

```python
# Fill missing numeric values with median, categorical with mode
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

# Remove duplicate rows if any
df.drop_duplicates(inplace=True)

print("✅ Data cleaned successfully!")
```

```python
# Add Profit (Revenue - Manufacturing - Shipping)
if {"Revenue generated", "Manufacturing costs", "Shipping costs"}.issubset(df.columns):
    df["Profit"] = (
        df["Revenue generated"] - df["Manufacturing costs"] - df["Shipping costs"]
    )

# Add OnTimeDelivery flag
if "Shipping times" in df.columns and "Lead time" in df.columns:
    df["OnTimeDelivery"] = np.where(
        df["Shipping times"] <= df["Lead time"], "Yes", "No"
    )

df.head()
```

`Profit` helps measure profitability per order.

`OnTimeDelivery` shows logistics efficiency — useful for KPI tracking.

# ***Exploratory Data Analysis (EDA)***

**Visualize inventory, sales, suppliers, and transport**

```python
plt.figure(figsize=(10,5))
inventory = df.groupby("Product type")["Availability"].mean().sort_values(ascending=False)
sns.barplot(x=inventory.index, y=inventory.values)
plt.title("Average Inventory Availability by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Average Availability")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**`Average Inventory Availability by Product Type`**

*  Haircare shows a lower average availability (about 43 units).

```python
if "Shipping carriers" in df.columns:
    transport = df.groupby("Shipping carriers")[["Shipping times", "Shipping costs"]].mean()
    fig = px.scatter(
        transport,
        x="Shipping times",
        y="Shipping costs",
        text=transport.index,
        size="Shipping costs",
        title="Carrier-wise Shipping Time vs Cost"
    )
    fig.show()
```

***Carrier-wise Shipping Time vs Cost Analysis***

*  Carrier B offers the lowest cost and shortest shipping time, making it the most efficient option.

# ***Business Insights & KPI Dashboard***
**Calculate metrics, display insights, and final report**

```python
total_revenue = df["Revenue generated"].sum()
total_costs = df["Manufacturing costs"].sum() + df["Shipping costs"].sum()
total_profit = total_revenue - total_costs
on_time_rate = df["OnTimeDelivery"].value_counts(normalize=True).get("Yes", 0) * 100

print("📈 SUPPLY CHAIN PERFORMANCE SUMMARY")
print("=====================================")
print(f"💵 Total Revenue: ₹{total_revenue:,.0f}")
print(f"💸 Total Costs: ₹{total_costs:,.0f}")
print(f"🏆 Total Profit: ₹{total_profit:,.0f}")
print(f"🚚 On-Time Delivery Rate: {on_time_rate:.2f}%")
print(f"📦 Avg Supplier Lead Time: {df['Lead time'].mean():.2f} days")
print(f"⚙️  Avg Defect Rate: {df['Defect rates'].mean():.2f}%")
print("=====================================")
```

```python
if "Profit" in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df["Profit"], bins=20, kde=True)
    plt.title("Profit Distribution Across Orders")
    plt.xlabel("Profit (₹)")
    plt.tight_layout()
    plt.show()

    # Cost Distribution Pie Chart
    cost_summary = {
        "Transportation": df["Shipping costs"].sum(),
        "Manufacturing": df["Manufacturing costs"].sum(),
        "Miscellaneous": df["Revenue generated"].sum() * 0.1
    }

    fig = px.pie(
        names=list(cost_summary.keys()),
        values=list(cost_summary.values()),
        title="Cost Distribution in Supply Chain"
    )
    fig.show()
```

**`Profit & Cost Distribution in Supply Chain`**


*   The histogram (top-left) shows profit distribution across orders, with peaks around 2,000–3,000 and 8,000–9,000, indicating varied profitability levels.

*   Profits appear bimodal, suggesting two distinct order segments or pricing structures.

*  The pie chart (bottom-right) illustrates cost distribution within the supply chain.
*   Miscellaneous costs dominate at over 90%, while manufacturing (≈7.5%) and transportation (≈0.9%) contribute relatively little.


* This indicates a high concentration of expenses in non-core or indirect areas, highlighting potential cost optimization opportunities.


```python
# ==============================================================
# FINAL SUPPLY CHAIN DASHBOARD VIEW
# ==============================================================

# Prepare key metrics
metrics = {
    "Total Revenue (₹)": df["Revenue generated"].sum(),
    "Total Costs (₹)": df["Manufacturing costs"].sum() + df["Shipping costs"].sum(),
    "Total Profit (₹)": df["Revenue generated"].sum() - (df["Manufacturing costs"].sum() + df["Shipping costs"].sum()),
    "Avg Supplier Lead Time (Days)": df["Lead time"].mean(),
    "Avg Defect Rate (%)": df["Defect rates"].mean(),
    "On-Time Delivery (%)": df["OnTimeDelivery"].value_counts(normalize=True).get("Yes", 0) * 100
}

# Convert to DataFrame
summary_df = pd.DataFrame(list(metrics.items()), columns=["Metric", "Value"])

# Interactive Plotly Bar + Line Chart
fig = px.bar(
    summary_df,
    x="Metric",
    y="Value",
    text="Value",
    title="📊 Overall Supply Chain Performance Dashboard",
    color="Metric",
)
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(
    title_font_size=20,
    title_x=0.5,
    xaxis_title="Business Metrics",
    yaxis_title="Value",
    showlegend=False,
    plot_bgcolor="#f5f5f5"
)
fig.show()
```

***Overall Supply Chain Performance Dashboard***


*  **Title Insight:** The dashboard represents the Overall Supply Chain Performance, focusing on business efficiency metrics.
*  Displayed Metric:The visible bar shows On-Time Delivery (%), a critical KPI for logistics and operations.


*   Y-Axis: Represents the value of the measured metric (around 102.057k), possibly cumulative or scaled data.
*  Purpose: To monitor and evaluate how consistently deliveries meet their deadlines.


*   Interpretation: The close values suggest data stability or minimal variation in performance across measured periods.







```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Calculate profit by product type
profit_by_product = df.groupby('Product type')['Profit'].sum().sort_values(ascending=False)

# Calculate profit contribution percentage
total_profit = df['Profit'].sum()
profit_contrib = (profit_by_product / total_profit * 100).sort_values(ascending=False)

# --------------------------------------------------------------
# Combined Dashboard View
# --------------------------------------------------------------
fig = make_subplots(rows=1, cols=2, subplot_titles=("Profit by Product", "Profit Contribution %"),
                    specs=[[{"type": "bar"}, {"type": "pie"}]])

fig.add_trace(
    go.Bar(x=profit_by_product.index, y=profit_by_product.values, marker_color='royalblue', name="Profit"),
    row=1, col=1
)
fig.add_trace(
    go.Pie(labels=profit_contrib.index, values=profit_contrib.values, name="Contribution", hole=0.3),
    row=1, col=2
)

fig.update_layout(
    title_text="💼 Overall Profit Perspective Dashboard",
    title_x=0.5,
    showlegend=False,
    height=500
)
fig.show()
```

**`Overall Profit Perspective Dashboard`**


*   Purpose: This dashboard shows how different product categories contribute to overall company profits.
*   Left Chart (Bar Graph): Displays Profit by Product — skincare leads (~₹240K), followed by haircare and cosmetics.

*   Right Chart (Donut Chart): Illustrates Profit Contribution % — skincare (41.8%), haircare (30.2%), and cosmetics (28%).
*   Insight: Skincare is the most profitable segment, making up nearly half of total profits.

*   Use Case: Helps management identify which product lines drive revenue and where to focus marketing or investment efforts.
